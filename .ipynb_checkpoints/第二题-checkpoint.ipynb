```python
import time
import pandas as pd
import numpy as np
from tqdm import tqdm

import tensorflow as tf
from transformers import AutoTokenizer, TFBertForSequenceClassification
from LAC import LAC

tf.__version__
```

```python
# 加载附件三数据 time.strptime(a1, "%Y-%m-%d %H:%M:%S")
data_df = pd.read_excel('./data/附件3.xlsx')
data_df['留言详情'] = data_df['留言详情'].apply(lambda x : x.strip())

# 按时间对数据进行排序
data_df['留言时间'] = pd.to_datetime(data_df['留言时间'])
data_df.sort_values('留言时间', inplace=True)
data_df.head()
```

```python
label2id = {'城乡建设': 0, '环境保护': 1, '交通运输': 2, '教育文体': 3, '劳动和社会保障': 4, '商贸旅游': 5, '卫生计生': 6}
id2label = {0: '城乡建设', 1: '环境保护', 2: '交通运输', 3: '教育文体', 4: '劳动和社会保障', 5: '商贸旅游', 6: '卫生计生'}

# 加载第一题训练好的模型
# test_data = tokenizer('A市经济学院强制16届电子商务跟企业物流专业实习。其中我们企业物流专业实习6个月', padding=True, truncation=True, max_length=380, return_tensors='tf')
bert_model_test = TFBertForSequenceClassification.from_pretrained('./bert_pretraind/', return_dict=True)
tokenizer = AutoTokenizer.from_pretrained('bert-base-chinese')

def build_bert_inputs(tokenizer, sentences):
    encoded_input = tokenizer(sentences, padding=True, truncation=True, max_length=380, return_tensors='tf')
    return encoded_input

sentences = list(data_df['留言详情'])
results = []
for sentence in tqdm(sentences):
    test_inputs = build_bert_inputs(tokenizer, sentence)
    # 构建预测函数
    outputs = bert_model_test(dict(test_inputs))
    logits = outputs.logits
    label_id = tf.argmax(tf.nn.softmax(logits)[0]).numpy()
    target = id2label[label_id]
    results.append(target)
len(results)
```

```python
# 保存预测结果，下次运行代码时直接加载，跳过上面的代码运行，节约时间
data_df['label'] = results
data_df.to_excel('./data/附件3预测一级标签.xlsx')
data_df.info()
```

```python
# 装载LAC模型
lac = LAC(mode='lac')
# 单个样本输入，输入为Unicode编码的字符串
text = u"LAC是个优秀的分词工具"
lac_result = lac.run(text)

# 批量样本输入, 输入为多个句子组成的list，平均速率更快
texts = data_df['留言详情'][0]
lac_result = lac.run(texts)
lac_result
```
